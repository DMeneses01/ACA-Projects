```python
y_trainv = train_df['class']
X_trainv = train_df.drop(columns=['class', 'filename', 'number'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_rem, y_train, y_rem = train_test_split(X_trainv, y_trainv, train_size=0.6, random_state=42, stratify=y_trainv)
X_valid, X_test, y_valid, y_test = train_test_split(X_rem, y_rem, test_size=0.5, random_state=42, stratify=y_rem)
```

```python
from sklearn.decomposition import PCA

pca_test = PCA(.85)
pca_test.fit(X_trainv)
print(pca_test.n_components_)
```

```python
pca_test = PCA(.90)
pca_test.fit(X_trainv)
print(pca_test.n_components_)
```

```python
pca_test = PCA(.95)
pca_test.fit(X_trainv)
print(pca_test.n_components_)
```

```python
def pca(data, target, n_components): 

    ppca = PCA(n_components)    
    principalComponents = ppca.fit_transform(data) 

    principalDf = pd.DataFrame(data = principalComponents, columns = ['Principal Component ' + str(i+1) for i in range(principalComponents.shape[1])])

    return principalDf
```

```python
pca_train = pca(X_train, y_train, 381)

display(pca_train)
```

```python
pca_valid = pca(X_valid, y_valid, 381)
```

### MLP sklearn

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import plot_confusion_matrix, confusion_matrix
from sklearn.metrics import classification_report

mlp = MLPClassifier()
mlp.fit(pca_train, y_train)

predict_test = mlp.predict(pca_valid)

print(confusion_matrix(y_valid, predict_test))

print(classification_report(y_valid, predict_test))
```
