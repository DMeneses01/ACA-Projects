```python
import numpy as np
import pandas as pd
import os
import cv2

# Define path to the dataset
PATH = "C:/Users/emanu/Universidade/Mestrado/1_Ano/2_Semestre/ACA/Projeto/ACA-Project/AML-Project2/dataset"

# Read the training set data
train_data = np.load(os.path.join(PATH, "trainX.npy"))
train_labels = np.load(os.path.join(PATH, "trainy.npy"), allow_pickle=True)

display(train_labels)

print(train_data.shape)
display(train_data)

# Read the test set data
test_data = np.load(os.path.join(PATH, "testX.npy"))

# Read the CSV file containing the image paths and labels
train_df = pd.read_csv(os.path.join(PATH, "train_intel-image-classification-csvdata.csv"))
test_df = pd.read_csv(os.path.join(PATH, "test_intel-image-classification-csvdata-kaggle.csv"))

display(train_df)

'''
# Display some sample images from the training set
for i in range(5):
    img = train_data[i]
    label = train_labels[i]
    label_name = train_df.loc[label]["class"]
    cv2.imshow(str(label_name), img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
'''
```

```python
y_trainv = train_df['class']
X_trainv = train_df.drop(columns=['class', 'filename', 'number'], axis=1)
```

```python
X_trainv = X_trainv / 255

'''
cols = X_trainv.columns

for c in cols:
    X_trainv[c] = (X_trainv[c] - X_trainv[c].mean()) / X_trainv[c].std()
'''

display(X_trainv)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_trainv, y_trainv, train_size=0.8, random_state=42, stratify=y_trainv)
```

### MLP sklearn

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import plot_confusion_matrix, confusion_matrix
from sklearn.metrics import classification_report

mlp = MLPClassifier(activation="logistic", solver="sgd", max_iter=100000, random_state=42)
mlp.fit(X_train, y_train)

predict_test = mlp.predict(X_test)

print(confusion_matrix(y_test, predict_test))

print(classification_report(y_test, predict_test))
```

```python
kaggle_df = test_df.drop(columns=['filename', 'number'], axis=1)

kaggle_df = kaggle_df / 255

'''
cols = kaggle_df.columns

for c in cols:
    kaggle_df[c] = (kaggle_df[c] - kaggle_df[c].mean()) / kaggle_df[c].std()
'''

display(kaggle_df)
```

```python
predict_kaggle = mlp.predict(kaggle_df)
```
